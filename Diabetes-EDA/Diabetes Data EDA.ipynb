### **4. Checking Missing Values**

- #### **isnull() -** Check null values in each column and display in boolean.

```python
df.isnull()
```

- #### **isnull().sum() -** Show total null values in each column.

```python
df.isnull().sum()
```

- #### **duplicated() -** Check duplicate row and display in boolean.

```python
df.duplicated()
```

- #### **duplicated().sum() -** Total count of duplicates row.

```python
df.duplicated().sum()
```

- #### **As there are no missing value, so we are checking for zeros in some columns, as zeros are not valid in medical data.**

```python
(df == 0).sum()
```

- #### **Some medical columns cannot be zero, so we replace them with NaN (missing value).**

```python
cols_with_zero = ["Glucose", "BloodPressure", "SkinThickness", "Insulin", "BMI"]

df[cols_with_zero] = df[cols_with_zero].replace(0, np.nan)
```

```python
(df == 0).sum()
```

- #### **We had replaced 0 with NaN, now again we count for missing values.**

```python
df.isnull().sum()
```

- #### **Replacing missing values(NaN) with median.**

- #### **fillna() -** Filling missing values with median.

```python
for col in cols_with_zero:
    df[col].fillna(df[col].median(), inplace=True)
```

```python
df.isnull().sum()
```

### **5. Identifying Outliers for each numerical column**

```python
numerical_cols = df.columns[:-1]
for col in numerical_cols:
    sns.boxplot(data=df, y=col, color="#e74c3c")
    plt.title(f"Outliers in {col}", 
              fontsize=15, fontweight="bold")
    plt.show()
```

### **5. Univariate Analysis**

- #### **Checking distribution of values for each column**.

```python
for col in numerical_cols:
    sns.histplot(
        data=df, 
        x=col, 
        color="green", 
        kde=True, 
        bins=20
    )
    plt.title(f"{col} Distribution", 
              fontsize=15, fontweight="bold")
    plt.show()
```

- #### **Outcome Conclusion :** Around 65% of individuals are non-diabetic and 35% of individuals are diabetic

```python
print(df["Outcome"].value_counts())
sns.countplot(
    data=df, 
    x="Outcome", 
    hue="Outcome", 
    palette=["#db5f57", "#36ada4"], 
    edgecolor="black"
)
plt.legend(
    title="Outcome",
    labels=["Non-Diabetes", "Diabetes"],
    loc="upper right"
)
plt.title("Outcome Distribution", 
          fontsize=15, fontweight="bold")
```

### **6. Bivariate Analysis -** Checking relationship between two variables

- #### **Pregnancies Vs Outcome Conclusion-** Women with diabetes has more pregnancies as comared to non-diabetic womens

```python
sns.barplot(
    data=df, 
    x="Outcome", 
    y="Pregnancies", 
    hue="Outcome", 
    palette=["#ff6f61", "#6b5b95"], 
    edgecolor="black"
)
plt.title("Pregnancies Vs Outcome", 
          fontsize=15, fontweight="bold")
plt.xticks([0, 1], ["0 : Non-Diabetes", "1 : Diabetes"])
```

- #### **Glucose Vs Outcome Conclusion-** Glucose level of a diabetic person is higher as compared to non-diabetic.

```python
sns.barplot(
    data=df,
    x="Outcome",
    y="Glucose", 
    hue="Outcome", 
    palette=["#ff6f61", "#6b5b95"], 
    edgecolor="black"
)
plt.title("Glucose Vs Outcome",
          fontsize=15, fontweight="bold")
plt.xticks([0, 1], ["0 : Non-Diabetes", "1 : Diabetes"])
```

- #### **Age Vs Outcome Conclusion-** As the age is increasing the probability of diabetes is also increasing.

```python
sns.barplot(
    data=df, 
    x="Outcome", 
    y="Age", 
    hue="Outcome",
    palette=["#ff6f61", "#6b5b95"], 
    edgecolor="black"
)
plt.title("Age Vs Outcome",fontsize=15, 
          fontweight="bold"
         )
plt.xticks([0, 1], ["0 : Non-Diabetes", "1 : Diabetes"])
```

- #### **Insulin Vs Outcome Conclusion-** Insulin level of a diabetic person is higher as compared to non-diabetic.

```python
sns.barplot(
    data=df, 
    x="Outcome", 
    y="Insulin", 
    hue="Outcome", 
    palette=["#ff6f61", "#6b5b95"],
    edgecolor="black"
)
plt.title("Insulin Vs Outcome",
          fontsize=15, fontweight="bold")
plt.xticks([0, 1], ["0 : Non-Diabetes", "1 : Diabetes"])
```

- #### **DiabetesPedigreeFunction Vs Outcome Conclusion-** Higher Diabetes Pedigree Function values are more likely to have diabetes.

```python
sns.barplot(
    data=df, 
    x="Outcome", 
    y="DiabetesPedigreeFunction",   
    palette=["#ff6f61", "#6b5b95"], 
    hue="Outcome", 
    edgecolor="black"
)
plt.title("DiabetesPedigreeFunction Vs Outcome",
          fontsize=15, fontweight="bold")
plt.xticks([0, 1], ["0 : Non-Diabetes", "1 : Diabetes"])
```

- #### **Creating boxplot with loop for bivariate analysis**
- #### **Numerical columns vs Outcome column**

```python
numerical_cols = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 
                  'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age']

for col in numerical_cols:
    sns.boxplot(
        data=df,
        x="Outcome",
        y=col, hue="Outcome",
        palette=['#2b8e70', '#f2b134'])
    plt.title(f'{col} vs Outcome', 
              fontsize=15, fontweight="bold")
    plt.xticks([0, 1], ["0 : Non-Diabetes", "1 : Diabetes"])
    plt.xlabel("Outcome")
    plt.ylabel(col)
    plt.show()
```

### **7. Multivariate Analysis -** Checking relationship between three or more variables

```python
sns.scatterplot(
    data=df, 
    x="Glucose", 
    y="BMI", 
    hue="Outcome", 
    palette = ['#2b8e70', '#e57f7f']
)
plt.legend(title="Outcome",
           labels=["Non-Diabetes", "Diabetes"],
           loc="upper left"
)
plt.title("Glucose vs BMI by Outcome", 
          fontsize=15, fontweight='bold')
```

```python
sns.scatterplot(
    data=df, 
    x="Age", 
    y="Pregnancies", 
    hue="Outcome", 
    palette = ['#2b8e70', '#e57f7f']
)
plt.legend(title="Outcome",
           labels=["Non-Diabetes", "Diabetes"],
           loc="upper right"
)
plt.title("Age vs Pregnancies by Outcome", 
          fontsize=15, fontweight='bold')
```

```python
sns.lineplot(
    data=df, 
    x="Age", 
    y="Glucose", 
    hue="Outcome", 
    palette = ['#2b8e70', '#e74c3c']
)
plt.title("Age vs Glucose by Outcome", 
          fontsize=15, fontweight='bold')
```

```python
vars=['Glucose', 'BMI', 'Age', 'Insulin', 'Outcome']
sns.pairplot(df, vars=vars, hue='Outcome', 
             palette = ['#2b8e70', '#e57f7f'])
```

```python
plt.figure(figsize=(8,6))
correlation = df.corr()
sns.heatmap(
    correlation, 
    annot=True, 
    cmap="BuGn", 
    fmt=".2f", 
    linewidths=0.5, 
    square=True)
plt.title("Correlation Matrix of Numeric Features", 
          fontsize=15, fontweight='bold')
plt.tight_layout()
plt.show()
```
